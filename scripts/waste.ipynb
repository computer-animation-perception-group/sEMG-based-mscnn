```python
%matplotlib inline
import matplotlib.pyplot as plt
import scipy.io as sio
import numpy as np
import seaborn as sns
sns.set_context('paper', rc={'font.family': 'Arial', 'font.size': 7})
sns.set_style('whitegrid')


def savefig(prefix):
    for ext in ['.png', '.jpg', '.eps', '.pdf']:
        plt.savefig(prefix + ext, dpi=300, bbox_inches='tight', pad_inches=0.05)


d = sio.loadmat('../.cache/dbb/001-101.mat')
data = d['data']
ch = data[:, 16]
plt.plot(ch)
plt.fill_between([7500, 9000], -0.4, 0.4, color='r', alpha=0.5)
plt.fill_between([7000, 11000], -0.4, 0.4, color='r', alpha=0.2)
plt.xlabel('Frames')
plt.ylim(-0.4, 0.4)
plt.xlim(6000, 12000)
savefig('../.cache/figures/waste')
```
